## If we want to do vector addition


there's two  ways 
- using regular lists
- using numpy arrays (faster)

```python
#normal way
q=[1,2,3,4]
w=[1,2,3,4]
z=[]
for n,m in zip(q,w):
   z.append(n+m) 

print(z)
```

```python
#numpy array
#it will run alot faster
q=np.array([1,2,3,4])
w=np.array([1,2,3,4])
z=q+w
print(z)
```

## Multiplication with a scalar


there's two ways
- normal way
- with numpy

```python
r=[1,2,4,5]
z=[]
for n in r:
    z.append(2*n)

print(z)
```

```python
r=np.array([1,2,4,5])
z=2*r
print(z)
```
