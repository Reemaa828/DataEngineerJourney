# Python Data Structure
## 1 Tuple (A Compound Data type)
- Tuple is an ordered sequence (meaning we can use an index)
- Tuple is written with a parenthese and a comma seperating each value
- it can contain different data types in the same tuple
- We can concatenate tuples together
- Tuple are **immutable**(you cannot change, add, or remove its elements)


- we can assign it to another tuple (now the tuple's name reference another tuple)



- we can sort it but store it in a different tuple
- Nesting in tuple is valid




```python
#Tuple syntax
tuple=(1,2,3,4,5,6)
print(type(tuple))

#Different data types is valid
tuple1=(1,1.0,'reem',33)

#Using index with tuple
print(tuple[0])
print(tuple[-1])

#Concatenate tuples
tuple3= tuple + tuple1
print(tuple3)

#Slicing 
print(tuple3[0:3])

#Length
print(len(tuple3))

#immutable
# tuple[0]=1 ,tuple.append(2), del tuple[2]
# but we can do this
tuple4=tuple # cause both are just references

#sort tuple
tuplereverse=sorted(tuple,reverse=True)
print(tuplereverse)

#Nesting
tuple5=(1,2,3,(4,5),6)
print(tuple5[3][1])
print(tuple5[3])


```

## 2 List (A Compound Data Type) 
- List is also an ordered sequence(we can use an index with it)
- List is represented by square brackets
- Lists are mutable (we can change, add and remove)
- we can nest and put together different types
- convert string to list using **split** 
-----------------------------------------------
- ALiasing (having two names of the same thing)



- changing one can lead for the change to show in another



### if i don't want an alias just a copy then do this
a=b[:]
*or*
a=list(b)



```python
#List syntax
l= ["reem",24,"engineering",7]
print(l[1])

#Nesting
ll=[1,2,3,4,[4,5,6],(6,7)]

#Index
print(l[2])

#Slicing
print(l[1:])

#concate
lll=ll+l
print(lll)

#append or add on list
lll.append(99)
print(lll)
lll.extend([2002,2020])
print(lll)

#change
lll[0]="change"
print(lll)

#delete
del lll[0]
print(lll)

# convert a string to a list using split
print("this is so cool".split())
print(r"reem\rer\reer\ere ".split('\\') ) #using another delimiter

#Aliasing
lll = ll
lll[0]="alias"
print(ll)
#copy
lll=ll[:]
lll[0]="no alaising"
print(ll) #old value didn't change

#Help using tuple or list
help(lll)


```

## 3 Sets
- Sets are unordered data type (we can't use index with it)
- Sets only have unique values
- Sets are represented with curly brackets {}
- we can convert list to set by using `set()`
- add to set `.add`
- remove from set `.remove`
- use `in` to check if the value is in the set
____
##### we need to think of sets as a venn diagram
- to **intersect** `&`




- to **union** `.union()`



- to check if it a **subset** of another set `<small set>.issubset(large set)`





```python
#Set syntax
s={"ree","rr","ee","r","e","rr"}
#Removes Duplicates
print(s)

#convert list to set
l=[1,3,2,3,4,5,"ree","rr","ee"]
ss=set(l)
print(ss)

#Add and remove data
s.add("qee")
print(s)
s.remove("qee")
s.add(8)
print(s)

#we need to think of sets as a Venn diagram
#in check
"ree" in s

#Intersect
s1= ss & s
print(s1)

#union
s1=s1.union(ss)
print(s1)

#subset
v={1,23,44,55,66,77}
vv={55,66}
print(vv.issubset(v))
```

## DIctionary 


- dictionary is represented by key value pair
- key must be unique and immutable(no adding, removing or changing)


```python
#Dictionary syntax
dic={"k112":433,"k1w2":"23ree","ewjej11":12.3}
print(dic)
print(dic["k112"])
print(dic.keys())
print(dic.values())
print(dic.items())

#Add an item
dic["1212"]="reem"
print(dic)

del dic["k112"]
print(dic)

#Check for an item
"1212" in dic
```
