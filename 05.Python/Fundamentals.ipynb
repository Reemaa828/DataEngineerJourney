# Python Fundamentals
## Comparison operators
- `==` equal sign
- `>` greater than
- `>=` greater than or equal
- `!=` not equal


````python
a=8
print(6==a)
print(8==a)
print(9>a)
print(8>=a)
````

## Loops
- Syntax of a for loop
`for <variable> in <iterable>:

- Syntax of a while loop
`while <conndition>:

````python
for x in range(2):
    print(x)
for x in range(4,9):
    print(x)
# ways to iterate between list
y =[1,2,3,4]
for x in y:
    print(x)

for x,z in enumerate(y):
    print(x,z)

for x in range(len(y)):
    print(y[x])

liss=[1,2,3,4,5]
liss1=[]
i=0
while liss[i] < 4:
    liss1.append(liss[i])
    i+=1
print(liss1)
````
